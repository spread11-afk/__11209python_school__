```python
import numpy as np
import matplotlib.pyplot as plt
# 提供 x 和 y 數據
x = np.array([x for x in range(0, 10)])
y = np.array([0,0.5,1.5,3.5,3.4,4.7,3.4,5,4.4,7])  # 這裡提供了一個示例 y 數據
# 繪製原始數據
plt.plot(x, y, marker='o', label="Data")
# 使用 polyfit() 擬合多項式
degree = 1  # 指定多項式的次數（例如，1代表一次直線）
coefficients = np.polyfit(x, y, degree)
# 使用 poly1d() 創建多項式對象
poly = np.poly1d(coefficients)
# 生成新的 x 值來計算趨勢線上的 y 值
x_fit = np.linspace(x[0], x[-1], 100)
y_fit = poly(x_fit)
# 繪製趨勢線
plt.plot(x_fit, y_fit, label="Trendline", color='red')
# 添加標籤和圖例
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
# 顯示圖表
# 計算 R 平方值
y_mean = np.mean(y)
y_pred = poly(x)
ss_total = np.sum((y - y_mean) ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r_squared = 1 - (ss_res / ss_total)
# 添加 R 平方值到圖表
plt.text(4,6, f"R-squared = {r_squared:.2f}")
# 顯示圖表
plt.show()
```
