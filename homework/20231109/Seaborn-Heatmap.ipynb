```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成一組隨機數，代表從常態分布中取樣的數據
mu, sigma = 0, 1  # 均值和標準差
data = np.random.normal(mu, sigma, 1000)

# 繪製概率密度曲線
sns.kdeplot(data, fill=True, color='blue', alpha=0.5)

# 添加標籤和標題
plt.xlabel('值')
plt.ylabel('概率密度')
plt.title('常態分布的概率密度曲線')

# 顯示圖形
plt.show()
```
